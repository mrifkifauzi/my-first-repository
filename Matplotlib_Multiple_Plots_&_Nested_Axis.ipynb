# **Pembuatan Subplot menggunakan Matplotlib**

**Import library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Persiapan Data**

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

**Membuat subplot**

```python
#membuat figure
fig, ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(18,8))

#subplot untuk (x & tan)
ax[0].plot(x, tan, color='b', label='tanX')
ax[0].set_title('Grafik tanx')

#subplot untuk (x & sin)
ax[1].plot(x, sin, color='r', label='sinX')
ax[1].set_title('Grafik sinx')

#subplot untuk (x & cos)
ax[2].plot(x, cos, color='g', label='cosX')
ax[2].set_title('Grafik cosx')

#add legend dan menampilkan plot
fig.legend(loc='upper center')
plt.show()
```

# **Membuat Subplot Mested Axis**

**Persiapan Data**

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

**Pembuatan Nested Axis**

```python
#membuat figure
fig, ax = plt.subplots(figsize=(13,8.3))

#untuk subplot (x & cos(x))
ax.plot(x, y)
ax.set_title('Grafik CosX', fontsize=10)

#untuk subplot (x & cos(x)) dan (x & cos(x**2))
ins1 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])  #posisi nested axis
ins1.plot(x, y2, color='r')
ins1.plot(x, y, color='b')
ins1.set_title('Grafik CosX^2', fontsize=10)

#untuk subplot (x & cos(x)) dan (x & cos(x**3))
ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])   #posisi nested axis
ins2.plot(x, y3, color='g')
ins2.plot(x, y, color='b')
ins2.set_title('Grafik CosX^3', fontsize=10)

#untuk subplot (x & cos(x)) dan (x & cos(x**4))
ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])    #posisi nested axis
ins3.plot(x, y4, color='black')
ins3.plot(x, y, color='b')
ins3.set_title('Grafik CosX^4', fontsize=10)

#untuk subplot (x & cos(x)) dan (x & cos(x**5))
ins4 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])   #posisi nested axis
ins4.plot(x, y5, color='y')
ins4.plot(x, y, color='b')
ins4.set_title('Grafik CosX^5', fontsize=10)

plt.show()
```
