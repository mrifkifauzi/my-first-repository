**Persiapan Library dan Data (x, y)**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

**Membuat Figure dan Axis**

```python
from termcolor import colored   #untuk membuat kalimat berwarna

#Membuat Figure dan Axis kosong tanpa data
print(colored('Figure dan Axis Kosong tanpa Data', 'blue', attrs=['bold']))
fig = plt.figure()
ax = fig.add_subplot()
plt.show()

#Membuat plot data x dan y secara Explicit
print(colored('\nPlot Data x dan y secara Explicit', 'blue', attrs=['bold']))
fig2 = plt.figure()
ax2 = fig2.add_subplot()
ax2.plot(x, y)
plt.show()

#Membuat plot data x dan y secara Implicit
print(colored('\nPlot Data x dan y secara Implicit', 'blue', attrs=['bold']))
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```
