**Import Library**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Persiapan Data untuk Plotting**

```python
x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

**Pembuatan plot dengan Matplotlib**

```python
#membuat figure
plt.figure(figsize=(9.8,6.5))

#multiple plot (x & cos), (x & cos3), (x & sin), (x & sin2)
plt.plot(x, cos, color='red', label='Cosx')
plt.plot(x, cos3, linestyle='dashed', color='green', label='3 Cosx')
plt.plot(x, sin, linestyle='dashdot', color='blue', label='Sin(x-90)')
plt.scatter(x, sin2, marker='x', s=20,  color='y', label='2 Sinx')

#Membuat legend, label (x & y), dan judul
plt.legend(loc='center', framealpha=0.48, fontsize=9)
plt.xlabel('X-Axis Label', fontsize=9)
plt.ylabel('Y-Axis Label', fontsize=9)
plt.title('Grafik Cos SIn')

#Menampilkan Plot
plt.show()
```
